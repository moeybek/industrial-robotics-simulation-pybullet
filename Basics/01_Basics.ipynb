```python
import pybullet
import pybullet_data
```

```python
phsics_client =pybullet.connect(pybullet.GUI) #we have GUI, DIRECT, SHARED_MEMORY, UDP, TCP
```

```python
#reset space to start
pybullet.resetSimulation()

#add paths for necessary data
pybullet.setAdditionalSearchPath(pybullet_data.getDataPath())

#set gravity
pybullet.setGravity(0,0,-9.81)
time_step = 1./240.
pybullet.setTimeStep(time_step)
```

```python
#load plane
plane_id = pybullet.loadURDF("plane.urdf")
```

```python
mass = 10 #kg
box_size = [0.5,0.5,0.5] #m
position = [2,0,0.3] #m
orientation = [0,0,0,1] #quaternion
collision_shape_id = pybullet.createCollisionShape(pybullet.GEOM_BOX, halfExtents=box_size) #create collision shape for box
visual_shape_id = pybullet.createVisualShape(pybullet.GEOM_BOX, halfExtents=box_size, rgbaColor=[1,0,0,1]) #create visual shape for box
box_id = pybullet.createMultiBody(mass, collision_shape_id, visual_shape_id, position, orientation) #combine collision and visual shapes into a single body
```
